```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# 성능 데이터
CLIP = [100.0, 100.0, 100.0, 100.0, 100.0, 100.0, 100.0, 100.0, 100.0, 100.0, 100.0, 100.0]
coop = [97.5, 78.6, 71.3, 70.7, 61.2, 68.5, 70.5, 81.2, 74.5, 60.4, 64.0, 72.5]
ours = [101.1, 99.9, 94.6, 99.1, 91.8, 99.6, 97.1, 98.6, 93.0, 96.2, 93.9, 96.8]
rpo = [100.2, 99.1, 98.8, 99.7, 90.7, 98.9, 95.2, 96.3, 89.0, 98.4, 92.7, 96.2]
cocoop = [100.5, 96.2, 88.6, 87.5, 80.6, 94.1, 74.3, 94.0, 89.7, 78.3, 86.6, 88.1]

# 데이터셋 및 모델 설정
datasets = ["Imagenet", "Caltech", "Pets", "Cars", "Flowers", "Food", "Aircraft", "Sun397", "DTD", "Eurosat", "UFC", "Average"]
models = ['CLIP', 'Ours', 'RPO', 'CoCoop', 'CooP']

# pandas DataFrame으로 데이터 변환
data = []
for idx, model in enumerate(models):
    model_data = [CLIP, ours, rpo, cocoop, coop][idx]
    for d_idx, dataset in enumerate(datasets):
        data.append({'Model': model, 'Dataset': dataset, 'Score': model_data[d_idx]})

df = pd.DataFrame(data)

# seaborn으로 막대 그래프 그리기
plt.figure(figsize=(15, 8))
sns.barplot(x='Dataset', y='Score', hue='Model', data=df, palette='viridis')

plt.xticks(rotation=45, ha="right")
plt.title('Scores by Model and Dataset with Seaborn', fontsize=16)
plt.ylim(60, 110)  # y축 범위 설정

plt.tight_layout()
plt.show()
```

```python
# CLIP = [float(x) for x in "100.0	100.0	100.0	100.0	100.0	100.0	100.0	100.0	100.0	100.0	100.0	100.0".split()]
coop=[float(x) for x in "97.5	78.6	71.3	70.7	61.2	68.5	70.5	81.2	74.5	60.4	64.0  72.5".split()]
ours = [float(x) for x in "101.1	99.9	94.6	99.1	91.8	99.6	97.1	98.6	93.0	96.2	93.9  96.8".split()]
rpo= [float(x) for x in "100.2	99.1	98.8	99.7	90.7	98.9	95.2	96.3	89.0	98.4	92.7 96.2".split()]
cocoop=[float(x) for x in "100.5	96.2	88.6	87.5	80.6	94.1	74.3	94.0	89.7	78.3	86.6 88.1".split()]
datasets =[x.split("/")[1] for x in "Source/Imagenet	Source/caltech	Source/pets	Source/cars	Source/flowers	Source/food	Source/aircraft	Source/sun397	Source/DTD	Source/eurosat	Source/ufc".split() ] +['Average']
data = np.array([ ours,rpo,cocoop,coop])
half_size = len(datasets) // 2
datasets_first_half = datasets[:half_size]
datasets_second_half = datasets[half_size:]

# 조정된 인덱스를 데이터셋의 반으로 나누기
adjusted_indices_first_half = adjusted_indices[:half_size]
adjusted_indices_second_half = adjusted_indices[half_size:] - adjusted_indices[half_size]
offset = bar_width * (len(models) - 1) / 2

# 데이터셋 및 모델 설정
models = ['Ours' , 'RPO', 'CoCoop','CooP']

# updated_colors 변수가 정의되지 않았으므로, 새로운 색상을 정의하고 코드를 다시 실행합니다.
updated_colors = ['#fd8d3c','#6baed6',  '#74c476', '#9e9ac8']

# 데이터셋을 반으로 나누어서 각각의 subplot에 그리기 - 색상 정의 후 재실행

# 두 개의 subplot 생성 (수직으로 2개)
fig, axs = plt.subplots(2, 1, figsize=(15, 16), sharey=True)

# 첫 번째 그래프 (상단) - 데이터셋의 첫 번째 절반
for i, model in enumerate(models):
    axs[0].bar(adjusted_indices_first_half + i * bar_width, data[i, :half_size], bar_width, label=model, color=updated_colors[i])

axs[0].set_xlabel('Dataset', fontsize=16)
axs[0].set_ylabel('Scores', fontsize=16)
axs[0].set_title('Cross Dataset Transfer Compared to CLIP Zero-shot (First Half)', fontsize=20)
axs[0].set_xticks(adjusted_indices_first_half + offset)
axs[0].set_xticklabels(datasets_first_half,  fontsize=16)
axs[0].set_ylim([60,105])
axs[0].axhline(y=100, color='r', linestyle='--', linewidth=2, label='Zeroshot')
axs[0].legend(fontsize=12)

# 두 번째 그래프 (하단) - 데이터셋의 두 번째 절반
for i, model in enumerate(models):
    axs[1].bar(adjusted_indices_second_half + i * bar_width, data[i, half_size:], bar_width, label=model, color=updated_colors[i])

axs[1].set_xlabel('Dataset', fontsize=16)
axs[1].set_ylabel('Scores', fontsize=16)
axs[1].set_title('Cross Dataset Transfer Compared to CLIP Zero-shot (Second Half)', fontsize=20)
axs[1].set_xticks(adjusted_indices_second_half + offset)
axs[1].set_xticklabels(datasets_second_half, fontsize=16)
axs[1].set_ylim([60,105])
axs[1].axhline(y=100, color='r', linestyle='--', linewidth=2, label='Zeroshot')
axs[1].legend(fontsize=12)

plt.tight_layout()
plt.show()
```
